```python
import pandas as pd

pd.set_option('display.max_columns', None)

# Read the CSV file located in the current directory
# The file name is 'turning_data_for_ai_simulations.csv'  
# df = pd.read_csv('c:/dev/hackaping_2025/input/labeled_data.csv')



df = pd.read_csv('turning_data_for_ai_simulations_items.csv')

# Display the first few rows of the DataFrame
df.head()
```

```python
# Check for null or missing values in the DataFrame
df.isnull().sum()
```

```python
# Genereate a summary of the DataFrame
df.describe(include='all')
```

```python
# Create histograms for all numeric columns in the DataFrame
df.hist(figsize=(20, 15), bins=50, xlabelsize=8, ylabelsize=8)
```

```python
import matplotlib.pyplot as plt

# Show scatter plot using Operation Time Per PC for X and and the Item ID for Y
plt.scatter(df['Operation Time Per PC'], df['Item Id'])
plt.xlabel('Operation Time Per PC')
plt.ylabel('Item ID')
plt.title('Scatter plot of Operation Time Per PC vs Item Id')
plt.show()

# List the outliers of the Operation Time Per PC column showing Item ID, Operation Time Per PC, Machine ID and Material Length
threshold = df['Operation Time Per PC'].quantile(0.90)
outliers = df[df['Operation Time Per PC'] > threshold]

# Sort the outliers by Operation Time Per PC in descending order
outliers = outliers.sort_values(by='Operation Time Per PC', ascending=False)
outliers[['Item Id', 'Operation Time Per PC', 'Machine Id', 'Material Length (mm)']].head(10)





```

```python
# Plot the total number of machines
df['Machine Id'].value_counts().plot(kind='bar')
plt.xlabel('Machine ID')
plt.ylabel('Number of Items')
```

```python
# Show the total amount of time in minutes spent on creating all items during a year on first 10 items. Columns are Item Id, Operation Time Per PC, Consumed Total 2024
ax = df[df['Item Id'].isin(range(1, 11))].assign(
    Total_Time_Spent=lambda x: x['Operation Time Per PC'] * x['Consumed Total 2024:']
).groupby('Item Id').agg({'Total_Time_Spent': 'sum'}).plot(kind='bar', ylabel='Total Time Spent', legend=False)

# Add the Consumed Total 2024 as numbers displayed vertically inside each bar
for bar, item_id in zip(ax.patches, range(1, 11)):
    consumed_total = df.loc[df['Item Id'] == item_id, 'Consumed Total 2024:'].sum()
    ax.text(bar.get_x() + bar.get_width() / 2, bar.get_height() / 2, f'Items {consumed_total:.0f}', 
            ha='center', va='center', rotation='vertical', fontsize=8, color='white')





```
